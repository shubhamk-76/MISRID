```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, metrics,transform
import cv2
```

```python
prediction = cv2.imread("../prediction10.png",cv2.IMREAD_GRAYSCALE)
gt = cv2.imread('../DRIVE/truth/10.png',cv2.IMREAD_GRAYSCALE)
width,height=prediction.shape
gt = cv2.resize(gt, (height, width))

```

```python
prediction.shape,gt.shape
```

```python
ssim_index = metrics.structural_similarity(prediction, gt)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.imshow(prediction, cmap='gray')
ax1.set_title('Prediction')

ax2.imshow(gt, cmap='gray')
ax2.set_title('Truth')

plt.suptitle(f'SSIM Index: {ssim_index:.2f}')
plt.show()
```

```python
def psnr(img1, img2):
    # if img1.shape != img2.shape:
    #     raise ValueError("Input images must have the same size")
    height,width=img2.shape

    mse = np.mean((img2 - img1) ** 2)
    # print(mse)
    # if mse == 0:
    #     return float('inf')  # Images are identical, PSNR is infinite
    # max_pixel = 255.0
    # psnr_value = 10 * np.log10(max_pixel**2 / mse)
    # height, width,_ = img1.shape
    # img2 = cv2.resize(img2, (width, height))
    # import numpy as np
    # import math
    # img1 = img1.astype(np.float64) / 255.
    # img2 = img2.astype(np.float64) / 255.
    # mse = np.mean((img1 - img2) ** 2)
    # psnr=10 * math.log10(1. / mse)   
    # print(f"The PSNR between the two images is: {psnr} dB")
    PSNR=10*np.log10((255*255)/mse)   
    return PSNR
```
